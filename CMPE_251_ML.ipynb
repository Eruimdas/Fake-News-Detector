## Machine Learning

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer 
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score # evaluation
from sklearn import preprocessing
import numpy as np
```

### Importing the Data

```python
# reading pickle files into the ml part.
x = pd.read_pickle('x.pickle')
y = pd.read_pickle('y.pickle')
x['alldata'] = x.Header + x.Data
```

```python
#creating an object called df which separates the alldata column from x dataframe
#this lines is written to control data size which will be used by the machine learning algorithms.
#smallest number must be 6600 for 1:1 Zaytung and Cumhuriyet separation
#First 3300 entries indicates only Zaytung news, which nearly all of the data provided by the website.
data_size = 6600
df = [" ".join(review) for review in x['alldata'][:data_size]]
y = y[:data_size]
```

```python
x
```

### Train Test Split

```python
#splitting the dataset into training and test set
#X_train, X_test, y_train, y_test = train_test_split(df, y, random_state=50, test_size = 0.25)
X_train, X_test, y_train, y_test = train_test_split(df, y, random_state=50, test_size = 0.25, stratify=y)
# creating train and test splits according to a random state of 50, with 1/4 test size.
```

### Vectorizing the text data which is splitted into X_train and X_test

```python
vectorizer = TfidfVectorizer()
train_vectors = vectorizer.fit_transform(X_train).toarray()
test_vectors = vectorizer.transform(X_test).toarray()
# vectorizing the train and test vectors and translating them into arrays.
```

```python
#X_train
```

### Making scaler transforms

```python
scaler = preprocessing.StandardScaler().fit(train_vectors)
X_train = scaler.transform(train_vectors)
X_test = scaler.transform(test_vectors)
# creating a scaler object and transforming train and test vectors into scaler form.
```

### Type Correction

```python
X_train = np.asarray(X_train,dtype=np.float64)
y_train = np.asarray(y_train,dtype=np.float64)
y_test = np.asarray(y_test,dtype=np.float64)
# making sure that vectors are in same known data type for machine learning algorithms, if not changing them to
# desired types.
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(X_train, y_train)
predictions = lr.predict(X_test)
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, predictions)
print(accuracy)
```

### Neural Network

```python
from sklearn.neural_network import MLPClassifier
clf = MLPClassifier(solver='lbfgs', alpha=1e-1, hidden_layer_sizes=(10, 10), random_state=1)
clf.fit(X_train, y_train) 

y_pred = clf.predict(X_test)
print(accuracy_score(y_test, y_pred))
```
